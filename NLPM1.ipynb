```python
paragraph = """
Natural Language Processing enables computers to understand human language.
The running dogs were barking loudly at the flying birds.
She is studying and writing better algorithms for processing textual data.
The children played happily in the beautiful gardens yesterday.
"""

print("Input Paragraph:")
print(paragraph)
```

```python
print("=" * 60)
print("TOKENIZATION")
print("=" * 60)

# --- NLTK Tokenization ---
nltk_word_tokens = word_tokenize(paragraph)
nltk_sent_tokens = sent_tokenize(paragraph)

print("\n[NLTK] Word Tokens:")
print(nltk_word_tokens)
print(f"\nTotal NLTK word tokens: {len(nltk_word_tokens)}")

print("\n[NLTK] Sentence Tokens:")
for i, sent in enumerate(nltk_sent_tokens, 1):
    print(f"  S{i}: {sent.strip()}")

# --- SpaCy Tokenization ---
doc = nlp(paragraph)

spacy_word_tokens = [token.text for token in doc if not token.is_space]
spacy_sent_tokens = [sent.text.strip() for sent in doc.sents]

print("\n[SpaCy] Word Tokens:")
print(spacy_word_tokens)
print(f"\nTotal SpaCy word tokens: {len(spacy_word_tokens)}")

print("\n[SpaCy] Sentence Tokens:")
for i, sent in enumerate(spacy_sent_tokens, 1):
    print(f"  S{i}: {sent}")
```

```python
print("=" * 60)
print("STEMMING  (NLTK only — SpaCy has no built-in stemmer)")
print("=" * 60)

porter   = PorterStemmer()
snowball = SnowballStemmer("english")

# Use only alphabetic tokens for meaningful comparison
alpha_tokens = [t for t in nltk_word_tokens if t.isalpha()]

print(f"\n{'Token':<20} {'Porter Stem':<20} {'Snowball Stem':<20}")
print("-" * 60)
for token in alpha_tokens:
    print(f"{token:<20} {porter.stem(token):<20} {snowball.stem(token):<20}")
```

```python
print("=" * 60)
print("LEMMATIZATION")
print("=" * 60)

# --- NLTK Lemmatization ---
lemmatizer = WordNetLemmatizer()

# POS tagging helps NLTK lemmatize correctly
def get_wordnet_pos(treebank_tag):
    from nltk.corpus import wordnet
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # default

tagged = pos_tag(alpha_tokens)
nltk_lemmas = [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in tagged]

print(f"\n{'Token':<20} {'NLTK Lemma':<20} {'SpaCy Lemma':<20} {'SpaCy POS'}")
print("-" * 70)

# --- SpaCy Lemmatization ---
spacy_data = {
    token.text: (token.lemma_, token.pos_)
    for token in doc
    if token.is_alpha
}

for (word, tag), nltk_lemma in zip(tagged, nltk_lemmas):
    spacy_lemma, spacy_pos = spacy_data.get(word, ("N/A", "N/A"))
    print(f"{word:<20} {nltk_lemma:<20} {spacy_lemma:<20} {spacy_pos}")
```

```python
print("=" * 60)
print("COMPARISON SUMMARY TABLE")
print("=" * 60)

rows = []
for (word, tag), nltk_lemma in zip(tagged, nltk_lemmas):
    porter_stem   = porter.stem(word)
    snowball_stem = snowball.stem(word)
    spacy_lemma, spacy_pos = spacy_data.get(word, ("N/A", "N/A"))
    rows.append({
        "Original"     : word,
        "Porter Stem"  : porter_stem,
        "Snowball Stem": snowball_stem,
        "NLTK Lemma"   : nltk_lemma,
        "SpaCy Lemma"  : spacy_lemma,
        "SpaCy POS"    : spacy_pos
    })

df = pd.DataFrame(rows).drop_duplicates(subset="Original")
pd.set_option('display.max_rows', 50)
pd.set_option('display.width', 120)
print(df.to_string(index=False))
```
