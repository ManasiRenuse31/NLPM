```python
from transformers import BertTokenizer, BertModel
from bertviz import head_view
import torch

# 1. Load Pre-trained Model and Tokenizer
# We use 'output_attentions=True' to ensure we can extract the scores
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name, output_attentions=True)

# 2. Define a sentence with interesting relationships
# "it" is an anaphor that refers back to "the animal"
sentence = "The animal didn't cross the street because it was too tired."

# 3. Prepare Inputs
inputs = tokenizer.encode(sentence, return_tensors='pt')
outputs = model(inputs)

# 4. Extract Attention Weights
# Attention shape: (num_layers, batch, num_heads, seq_len, seq_len)
attention = outputs.attentions
tokens = tokenizer.convert_ids_to_tokens(inputs[0])

# 5. Render Visualization
# NOTE: This will display an interactive widget in your Jupyter/Colab cell
head_view(attention, tokens)
```
